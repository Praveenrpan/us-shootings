```python
import pandas as pd
mother_jones_file = '/Users/dgollapally/other-projects/us-shootings/resources/MotherJonesUsShootingData1982-present.csv'
nbc_file = '/Users/dgollapally/other-projects/us-shootings/resources/NBC.csv'
mj_df = pd.read_csv(mother_jones_file)
mj_df = mj_df.drop(columns=['Total victims','Prior signs of mental health issues',
                          'Mental health - details','Weapons obtained legally',
                          'Where obtained','Type of weapons','Weapon details',
                          'Race','Gender','Sources','Mental Health Sources',
                          'Case','Summary','Type','Venue','latitude','longitude',
                          'Injured','Location', 'Date'])
nbc_df = pd.read_csv(nbc_file).drop(columns=['Injured','Location', 'Date'])
df = pd.concat([mj_df, nbc_df])
df_groupby = df.groupby(['Year']).agg(['sum'])
ax = df_groupby.plot(legend=False,figsize=(15,4))
ax.set_ylabel("Fatalities")
ax
```
